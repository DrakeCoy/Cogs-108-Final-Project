### Cleaning up Dates

```python
#helper functions to clean up date columns
def year_from_date(date_time):
    year = date_time.split('-')[0]
    return int(year)

def month_from_date(date_time):
    month = date_time.split('-')[1]
    return int(month)

def day_from_date(date_time):
    date = date_time.split(' ')[0]
    day = date.split('-')[2]
    return int(day)

def date_from_date(date_time):
    date = date_time.split(' ')[0]
    return date

def time_from_date(date_time):
    date = date_time.split(' ')[1].split('+')[0]
    return date
```

**High** - keeping - The highest price of the cryptocurrency for a given day.  
**Low** - keeping - The lowest price of the cryptocurrency for a given day.  
**Close** - keeping - The price of the cryptocurrency when the market closes for the day.  
**Adj Close** - keeping - Stands for adjusted closing price. The adjusted closing price amends a stock's closing price to reflect that stock's value after accounting for any corporate actions. These appear to be the same as the normal closing price (Close).  
**Volume** - keeping - The amount of the cryptocurrency traded. More specifically, the volume is the value of all transactions for the given day.

### Example
I want to start by understanding the process of basic sentiment analysis on a single tweet.

Our tweet: "@spacex360 So great to see the happy faces!" This seems quite positive, hopefully VADER will think so.

```python
#example: 1 tweet; this is the tweet we'll use
print(elon_musk_tweets.loc[1,'content'])
```

```python
#example: 1 tweet
tokenized_tweet = word_tokenize(elon_musk_tweets.loc[1,'content'])
tokenized_tweet
#interesting, it splits the punctuation (!, @) into their own tokens, too.
```

```python
#set stop words
stop_words = set(stopwords.words('english'))
stop_words
#note we may want to find a way to put some other stop words in here. We probably don't need the @'s, for example'
```

```python
#remove stop words
for w in tokenized_tweet:
    if w in stop_words:
        tokenized_tweet.remove(w)
print(tokenized_tweet)
#it worked!
```

```python
#get stems for all words
stemmed_tweet = []
for w in tokenized_tweet:
    stemmed_tweet.append(ps.stem(w))
stemmed_tweet
```

```python
#sentiment analysis!
#note you can't get the polarity score of a list itself; you have look at every individual item
#but analyzing a cleaned sentence would just give you something like this
analysed_tweet = []
for w in stemmed_tweet:
    analysed_tweet.append(analyser.polarity_scores(w))
analysed_tweet
```

```python
#This is what confuses me, though:
#I can just analyse the entire text of the tweet
#so what is the point of doing the rest of everything that I did?
analyser.polarity_scores(elon_musk_tweets.loc[1,'content'])
#at least this looks quite positive
```

```python
#one last question: does the @ matter?
test1 = analyser.polarity_scores('So great to see the happy faces!')
test2 = analyser.polarity_scores('hahaha So great to see the happy faces!')
test3 = analyser.polarity_scores('@hahaha So great to see the happy faces!')
print(test1['compound'], test2['compound'], test3['compound'])
#no, it seems that the @ doesn't matter
```

#### What we see so far:
- It doesn't seem like all the work that was done to tokenize and remove words from the original sentences actually does anything for sentiment analysis. At the very least, the polarity scores function doesn't play nicely with it. Perhaps this is worth asking about.
- Vader ignores anything with an @ attached to it, which is kind of nice for not having to split up words.

### Actual Sorting

With everything I've just learned, it seems like best course of action is just to analyse all of the sentences. Then we can use the other text stuff above, such as tokenization, if we want to look for specific words (and maybe other uses?).

```python
def get_sentiments(input_list):
    
    output = pd.DataFrame()

    for sentence in input_list:
        ss = analyser.polarity_scores(sentence)
        ss['sentence'] = sentence
        output = output.append(ss, ignore_index=True)

    return output
```

```python
#note that this takes quite a while, ~15-30 seconds, to run.
elon_tweets_list = list(elon_musk_tweets['content'].values)
elon_sentiments = get_sentiments(elon_tweets_list)
```

```python
#Method 2: labeling/binning:
#We can sort the tweets into one of three categories based on their compound sentiment: positive, negative, or neutral.
#I'll work to see what works best for cutoffs, but I'm going to start with cutoffs around 0.3:

def bin_sentiments(compound):
    #This function aims to sort tweets into one of three categories based on their compound sentiment:
    
    #negative: tweets with a compound sentiment of -0.25 or lower
    #neutral: tweets with a compound sentiment between -0.25 and 0.25
    #positive: tweets with a compound sentiment of 0.25 or higher
    
    #I made this decision after looking over a few tweets in the range of 0.2-0.4 on each side and deciding 0.25 was a fair margin.
    
    if compound <= -0.25:
        return 'negative'
    
    if -0.25 < compound < 0.25:
        return 'neutral'
    
    if compound >= 0.25:
        return 'positive'
```

## Autoregressive Integrated Moving Average Model (ARIMA) 
The ARIMA model is a class of models that is used for analyzing and forecasting data over time. To help with finding lag for our model, we will use only recent data (March 2021 onward). For this model, we will analyze bitcoin price. This was adapted from an article on machinelearningmaster.com, titled "How to Create an ARIMA Model for Time Series Forecasting in Python," written by Jason Brownlee.

```python
series = btc_data[['date','close']]
series = series[series['date'] >= '2021-03-01']
series['date'] = series['date'].apply(pd.to_datetime)
series = series.set_index('date')

pd.plotting.autocorrelation_plot(series)
series.plot()
```

As shown in the autocorrelation plot, the autocorrelation is within absolute value of 0.3. Looking at lag, we can start by selecting a parameter for the AR model. We will start off with 5. The order parameter for the ARIMA function will be in order p, d, and q. This represents the number of lag observations, the number of times the raw observations are differenced, and the size of moving average window, respectively. 

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(series, order=(5,1,0))
model_fit = model.fit()

print(model_fit.summary())
```

Now we want to examine the residuals.

```python
residual_series = pd.DataFrame(model_fit.resid).reset_index()

plt.scatter(residual_series['date'],residual_series[0])
```

Based on the spread residuals, we see that our regression line does not need any transformation. Now we can move onto predicting the values. Our train,test split will have 0.8 and 0.2 for training and testing respectively.

```python
series.index = series.index.to_period('M')
```

```python
X = series.close
X_train, X_test = X[:int(0.8*X.shape[0])],X[int(0.8*X.shape[0]):]


preds = []
for i in range(X_test.shape[0]):
    mdl = ARIMA(X_train, order = (5,1,0))
    mdl_fit = mdl.fit()
    output = model_fit.forecast()
    preds.append(output)
    
```

```python

# split into train and test sets
X = series.values
size = int(len(X) * 0.6)
X_train, X_test = X[0:size], X[size:len(X)]
hist = [x for x in X_train]

preds = []

# walk-forward validation
#adapted from article
for i in range(len(X_test)):
    model = ARIMA(hist, order=(6,1,0))
    model_fit = model.fit()
    output = model_fit.forecast()
    pred = output[0]
    preds.append(pred)
    obs = X_test[i]
    hist.append(obs)

rmse = np.sqrt(((preds - X_test)** 2).mean())

plt.plot(X_test, color = 'red')
plt.plot(preds, color='green')
print(f'rmse = {rmse}')
```

As we can see above, there is a slight shift to the right of the predicted values. However, the ARIMA model forecasts the prices to be relatively similar to the actual, expected prices. The rmse, however, displays that there is a large difference between the predicted and expected.
